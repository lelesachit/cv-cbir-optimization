```python
import cv2

def load_image(file_path):
    # Load the image
    image = cv2.imread(file_path)
    # Convert the image from BGR to RGB
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image

# Load the image
image = load_image("sample.jpg")
```
