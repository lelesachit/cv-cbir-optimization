```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from preprocess import preprocessor

def blend_images(original_image, alpha):
    preprocessed_image = preprocessor(original_image)
    blended_image = cv2.addWeighted(original_image, alpha, preprocessed_image, 1-alpha, 0)
    return preprocessed_image, blended_image

# Example usage
original_image = cv2.imread('sample2.jpg')
alpha_values = [0.2, 0.4, 0.6, 0.8]

# Create a 2x3 subplot layout
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Plot the original image
axes[0, 0].imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
axes[0, 0].set_title('Original Image')
axes[0, 0].axis('off')

# Plot the returned image from the preprocessor function
preprocessed_image = preprocessor(original_image)
axes[0, 1].imshow(preprocessed_image, cmap='gray')
axes[0, 1].set_title('Preprocessed Image')
axes[0, 1].axis('off')

# Plot the blended images with different alpha values
for i, alpha in enumerate(alpha_values, start=2):
    _, blended_image = blend_images(original_image, alpha)
    axes[i // 3, i % 3].imshow(cv2.cvtColor(blended_image, cv2.COLOR_BGR2RGB))
    axes[i // 3, i % 3].set_title(f'Blended Image (Alpha={alpha})')
    axes[i // 3, i % 3].axis('off')

plt.tight_layout()
plt.show()
```
